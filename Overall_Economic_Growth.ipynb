```python
import pandas as pd
import numpy as np
import statistics as stat

# https://jakubmarian.com/how-to-understand-gdp-indicators/
# Difference between GDP indicators
```

```python
#Which five countries have experienced the greatest economic growth?
def calculate_gdp_percent_change(row, initialYear, finalYear):
    return (((row[finalYear] - row[initialYear]) / row[initialYear]) * 100)

gdp_columns_US_current = pd.read_csv('GDPPerCapita_Current_US$')
gdp_columns_US_current = gdp_columns_US_current[(gdp_columns_US_current['1991'].notnull()) &\
                                          (gdp_columns_US_current['1995'].notnull()) &\
                                          (gdp_columns_US_current['2000'].notnull()) &\
                                          (gdp_columns_US_current['2005'].notnull()) &\
                                          (gdp_columns_US_current['2010'].notnull())\
                                          & (gdp_columns_US_current['2017'].notnull())]

gdp_columns_US_current['gdpPercentChange'] = gdp_columns_US_current.apply(lambda row: calculate_gdp_percent_change(row, '1991', '2017'), axis=1)
gdp_columns_US_current = gdp_columns_US_current.sort_values(by=['gdpPercentChange'], ascending=False)

gdp_1991_to_2017_change = gdp_columns_US_current[['Country Name', 'gdpPercentChange']]
gdp_1991_to_2017_change.head(5)
```
